Clean up the merge data

```python
mergeGB = mergeTK.groupby("InputSize")

mergeCtk = th.Thicket.concat_thickets(
    thickets=list(mergeGB.values()),
    headers=list(mergeGB.keys()),
    axis="columns",
    metadata_key="num_procs"
)

mergeCtk.dataframe = mergeCtk.dataframe.reset_index().drop([("node", ""), (1024, "nid"), (1024, "spot.channel"), 
                                                            (1024, "Total time")], axis=1)
```

```python
radixMain = radixCtk.dataframe.loc[radixCtk.dataframe['name'] == 'main']
mergeMain = mergeCtk.dataframe.loc[mergeCtk.dataframe['name'] == 'main']

algosMain = {'Merge': mergeMain, 'Radix': radixMain}
```

```python
def plot_times(algos, num_vals, time_type, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    for k in algos.keys():
        dframe = algos[k]
        xvals = dframe['num_procs']
        yvals = dframe[num_vals][time_type]
        ax.plot(xvals, yvals, label=k)
    ax.legend()
    plt.xscale("log", base=2)
    plt.show()
```

```python
plot_times(algosMain, 1024, 'Avg time/rank', 'Strong Scaling: Input Size 1024', '# of processors', 'Time (s)')
```

```python
def plot_memcpy(df1, df2, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.plot(df1[['num_threads']], df1[['cudaMemcpy_host_to_device_time']], '-o', color='b', label='host to device')
    ax.plot(df2[['num_threads']], df2[['cudaMemcpy_device_to_host_time']], '-x', color='r', label ='device to host')
    ax.legend()
    plt.xscale("log", base=2)
    plt.show()

def plot_calc(df1, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.plot(df1[['num_threads']], df1[['bitonic_sort_step_time']])
    plt.xscale("log", base=2)
    plt.show()

def plot_band(df1, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.plot(df1[['num_threads']], df1[['effective_bandwidth (GB/s)']])
    plt.xscale("log", base=2)
    plt.show()

for num_values in test.groups.keys():
    rows_with_vals = meta.loc[test.groups[num_values]].sort_values('num_threads')
    htod = rows_with_vals[['num_threads', 'cudaMemcpy_host_to_device_time']]
    dtoh = rows_with_vals[['num_threads', 'cudaMemcpy_device_to_host_time']]
    plot_memcpy(htod, dtoh, 'Memcpy with ' + str(num_values) + ' Values', 'Threads', 'Time (ms)')

for num_values in test.groups.keys():
    rows_with_vals = meta.loc[test.groups[num_values]].sort_values('num_threads')
    calc = rows_with_vals[['num_threads', 'bitonic_sort_step_time']]
    plot_calc(calc, 'Bitonic Sort Step with ' + str(num_values) + ' Values', 'Threads', 'Time (ms)')

rows_with_vals = meta.loc[test.groups[16777216]].sort_values('num_threads')
band = rows_with_vals[['num_threads', 'effective_bandwidth (GB/s)']]
plot_band(band, 'Effective Bandwidth vs. Thread Count with 2^24 values', 'Threads', 'Effective Bandwidth (GB/s)')

# def plot_mpl(df, title, xlabel, ylabel):
#     fig = plt.figure()
#     ax = fig.add_subplot(1, 1, 1)
#     ax.plot(
#         df,
#         marker='o',
#     )
#     ax.set_title(title)
#     ax.set_xlabel(xlabel)
#     ax.set_ylabel(ylabel)
#     ax.legend(df.columns)
#     plt.grid(True)
#     plt.yscale("log", base=10)
#     plt.xscale("log", base=2)
#     plt.xticks(threads)
#     plt.show()
#     return fig, ax
```

# Plotting

### Option 1
Use `pandas.DataFrame.plot` to automatically interface with matplotlib

```python
def plot_pandas(df, title, xlabel, ylabel, loglog):
    plot = df.plot(kind="line", 
        marker='o', 
        title=title, 
        xlabel=xlabel,
        ylabel=ylabel, 
        grid=True,
        loglog=loglog
    )
    return plot
```

### Option 2
Use `matplotlib.pyplot` directly for more fine-grained control. Specifically for our case we can change the x axis scaling from base 10 to base 2.

```python
def plot_mpl(df, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o',
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(df.columns)
    plt.grid(True)
    plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax
```

```python
print(function_names)

ddFrames = []

for val in [128, 1024, 8192]:
    whole_comp=ctk.dataframe.loc['whole_computation']
    whole_comp=whole_comp.xs(val, level=0, axis=1)
    whole_comp=whole_comp['Avg time/rank']

    master_init=ctk.dataframe.loc['master_initialization']
    master_init=master_init.xs(val, level=0, axis=1)
    master_init=master_init['Avg time/rank']

    master_recv=ctk.dataframe.loc['master_send_recieve']
    master_recv=master_recv.xs(val, level=0, axis=1)
    master_recv=master_recv['Avg time/rank']
    result = pd.concat([whole_comp, master_init, master_recv], keys=['whole_computation', 'master_initialization', 'master_send_recieve'], axis=1)
    ddFrames.append(result)

plot_mpl(ddFrames[0], 'Matrix Size 128x128', "Processors", "Time (sec)")
plot_mpl(ddFrames[1], 'Matrix Size 1024x1024', "Processors", "Time (sec)")
plot_mpl(ddFrames[2], 'Matrix Size 8192x8192', "Processors", "Time (sec)")
```
