```python
import sys
from glob import glob

import matplotlib.pyplot as plt
import pandas as pd

import thicket as th

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
sorted_type = "random"
algorithm = "Radix"
implementation = "CUDA"
# file_path = "../MergeSort/CUDA/dataFiles/random/*.cali"

file_path = f"dataFiles/{implementation}/{algorithm}/{sorted_type}/*.cali"
# file_path = f"dataFiles/{implementation}/{algorithm}/*.cali"
print(file_path)

all_files = glob(file_path)
print(all_files)

tk = th.Thicket.from_caliperreader(all_files)
```

```python
print(tk.dataframe.keys())
```

```python
tk.metadata
```

```python

gb = tk.groupby("InputSize")
print(gb.keys())
```

```python
ctk = th.Thicket.concat_thickets(
  thickets=list(gb.values()),
  headers=list(gb.keys()),
  axis="columns",
  metadata_key="num_threads"
)
```

```python
ctk.dataframe
```

```python
desiredCol = "Total time"
sizes = [65536, 262144, 1048576, 4194304, 16777216, 67108864, 268435456]

desiredCols = [desiredCol, 'name']

# dropTheseCol = ["nid", "spot.channel",  "Min time/rank", "Max time/rank", "Avg time/rank", "Avg GPU time/rank" '']
dropTheseCol = []

# drop every column except 
keys = tk.dataframe.keys()
for k in keys:
    if k not in desiredCols:
        dropTheseCol.append(k)

print(dropTheseCol)

dropArr = [("node", "")]

for c in dropTheseCol:
    for s in sizes:
        dropArr.append((s, c))
    

ctk.dataframe = ctk.dataframe.reset_index().drop(dropArr, axis=1)

ctk.dataframe = ctk.dataframe.rename({("name", ""): "name",("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])


ctk.dataframe
```

```python
function_names = set(ctk.dataframe.index.get_level_values("name"))
world_size = sorted(list(set(ctk.dataframe.index.get_level_values("num_threads"))))
num_values = sorted(ctk.dataframe.columns.tolist())

print(function_names)
```

```python
def plot_pandas(df, title, xlabel, ylabel, loglog):
    plot = df.plot(kind="line",
                  marker='o',
                  title=title,
                  xlabel=xlabel,
                  ylabel=ylabel,
                  grid=True,
                  loglog=loglog
                )
    return plot
```

```python
ctk.dataframe.loc['main']
```

```python
for name in function_names:
    plot_pandas(
        df=ctk.dataframe.loc[name],
        title=f"{implementation} {algorithm} {sorted_type} input: {name}",
        xlabel="Threads",
        ylabel="Time (seconds)",
        loglog=True
    )
```
